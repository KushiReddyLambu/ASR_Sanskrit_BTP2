```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
import os
import shutil
```

```python
def convert_audio(audio_file, target_format="wav"):
    """
    Corrects the channels, sample rate, and sample width of the audios.
    Replaces the original audio file with the one generated.
    """
    try:
        sound = AudioSegment.from_file(audio_file)
        sound = sound.set_frame_rate(16000)
        sound = sound.set_channels(1)
        sound = sound.set_sample_width(2)  # 2 corresponds to 16-bit sample width in Pydub
        new_audio_file = audio_file.rsplit('.', 1)[0] + '.' + target_format
        sound.export(new_audio_file, format=target_format)
        return new_audio_file
    except Exception as e:
        print(f"Error in converting audio: {e}")
        return None
```

```python
# Step 3: Adjust the sampling rate, sample width, and channels
converted_audio = convert_audio("Atharvaveda Part 018.mp3")
```

```python
if converted_audio:
    sound_file = AudioSegment.from_file(converted_audio)
    audio_chunks = split_on_silence(sound_file,
                                    min_silence_len=4,
                                    keep_silence=500,
                                    silence_thresh=-40)
```

```python
files_and_directories = os.listdir()

for item in files_and_directories:
    print(item)
```

```python
# Directory where the chunks will be saved
output_dir = "chunked_files_13,14"

# Step 1: Remove the directory if it exists
if os.path.exists(output_dir):
    shutil.rmtree(output_dir)

# Step 2: Create the directory
os.makedirs(output_dir)
```
